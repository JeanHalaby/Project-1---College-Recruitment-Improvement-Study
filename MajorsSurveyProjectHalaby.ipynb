```python
# Data Cleaning Part
```

```python
# Renaming Columns
df = df.rename(columns={"race/ethnicity": "race",
                        "which_course_are_you_enrolled_in?": "course",
                       "how_did_you_hear_about_county_college_of_morris?_[ccm_web_site]": "heard_ccm_web",
                        "how_did_you_hear_about_county_college_of_morris?_[social_media]": "heard_ccm_socials",
                        "how_did_you_hear_about_county_college_of_morris?_[community_event]": "heard_ccm_community",
                        "how_did_you_hear_about_county_college_of_morris?_[family_member_or_friend]": "heard_ccm_fam_friend",
                        "how_did_you_hear_about_county_college_of_morris?_[current_ccm_student]": "heard_ccm_student",
                        "how_did_you_hear_about_county_college_of_morris?_[ccm_alumni]": "heard_ccm_alumni",
                        "how_did_you_hear_about_county_college_of_morris?_[high_school_teacher]": "heard_ccm_hs_teach",
                        "how_did_you_hear_about_county_college_of_morris?_[high_school_counselor]": "heard_ccm_hs_counselor",
                       "did_you_receive_information_about_the_ccm_computing_programs_from_any_of_the_following_sources?_[employer]": "comp_info_employer",
                        "did_you_receive_information_about_the_ccm_computing_programs_from_any_of_the_following_sources?_[ccm_workforce_development]": "comp_info_workforce",
                        "did_you_receive_information_about_the_ccm_computing_programs_from_any_of_the_following_sources?_[nj_workforce_development_program]": "comp_info_njworkforce",
                        "did_you_receive_information_about_the_ccm_computing_programs_from_any_of_the_following_sources?_[other]": "comp_info_other",
                       "was_a_computing_major/certificate_your_first_choice,_or_did_you_change_majors_from_a_different_ccm_program?_if_you_changed_majors,_indicate_what_your_first_major_was.": "first_choice",
                       "on_a_scale_of_1_to_5,_with_1_being_not_at_all_interested_and_5_being_extremely_interested,_how_interested_are_you_in_taking_more_computing_classes?": "scale_rating",
                        "please_explain_your_answer_to_the_question_above.__why_or_why_not_would_you_be_interested_in_taking_another_computing_class?": "reason"
                        })
df.head()
```

```python
# Cleaning up race
df['race'] = df['race'].replace({'Black/African American': 'Black' ,
                                                     'White/Caucasian': 'White' ,
                                                     'Hispanic or Latino': 'Hispanic',
                                                     'American Indian/Native American/Alaska Native': 'Native American'})
df.head()
```

```python
# Filling in NaN
df = df.fillna("Unknown")
df.head()
```
